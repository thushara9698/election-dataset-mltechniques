```python
#Logistic regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix

#X=data.drop(['WINNER','ASSETS','LIABILITIES','GENVOTES','AGE','POSTVOTES','TOTVOTES','ELECTORS_CONSTITUENCY','VOTES_CONSTITUENCY'],axis=1)
X=data.drop(['STATE','CONSTITUENCY','NAME','GENDER','CRCASES','PARTY','SYMBOL','EDUCATION','CATEGORY','WINNER','ASSETS','LIABILITIES'],axis=1)
y= data['WINNER']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)

lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
#print(classification_report(y_test,y_pred))
lr1=(accuracy_score(y_test,y_pred))
lr2=(classification_report(y_test,y_pred))
lr3=(lr.score(X_train,y_train))
lr4=(lr.score(X_test,y_test))
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
nb1=(accuracy_score(y_test,y_pred))
nb2=(classification_report(y_test,y_pred))
nb3=(nb.score(X_train,y_train))
nb4=(nb.score(X_test,y_test))
```

```python
#Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier
sgd=SGDClassifier(loss='modified_huber',shuffle=True,random_state=101)
sgd.fit(X_train,y_train)
y_pred=sgd.predict(X_test)
sgd1=(accuracy_score(y_test,y_pred))
sgd2=(classification_report(y_test,y_pred))
sgd3=(sgd.score(X_train,y_train))
sgd4=(sgd.score(X_test,y_test))
```

```python
#K Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
knn1=(accuracy_score(y_test,y_pred))
knn2=(classification_report(y_test,y_pred))
knn3=(knn.score(X_train,y_train))
knn4=(knn.score(X_test,y_test))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=10,random_state=101,max_features=None,min_samples_leaf=15)
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
dt1=(accuracy_score(y_test,y_pred))
dt2=(confusion_matrix(y_test,y_pred))
dt3=(dt.score(X_train,y_train))
dt4=(dt.score(X_test,y_test))
```

```python
#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rfm=RandomForestClassifier(n_estimators=40,oob_score=True,n_jobs=1,random_state=101,max_features=None,min_samples_leaf=15)
rfm.fit(X_train,y_train)
y_pred=rfm.predict(X_test)
rfm1=(accuracy_score(y_test,y_pred))
rfm2=(classification_report(y_test,y_pred))
rfm3=(rfm.score(X_train,y_train))
rfm4=(rfm.score(X_test,y_test))
```

```python
#Support Vector Machine
from sklearn.svm import SVC
svm=SVC(kernel="linear",C=0.025,random_state=101)
svm.fit(X_train,y_train)
y_pred=svm.predict(X_test)
svm1=(accuracy_score(y_test,y_pred))
svm2=(classification_report(y_test,y_pred))
svm3=(svm.score(X_train,y_train))
svm4=(svm.score(X_test,y_test))
```
